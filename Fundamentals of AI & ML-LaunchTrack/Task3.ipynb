```python
import random

def coin_toss_simulation(n_tosses):
    results = []
    heads_count = 0
    probs = []

    for i in range(1, n_tosses + 1):
        toss = random.choice(['Heads', 'Tails'])
        results.append(toss)
        if toss == 'Heads':
            heads_count += 1
        probs.append(heads_count / i)

    return results, probs

# Simulate 1000 tosses
n_trials = 1000
results, probs = coin_toss_simulation(n_trials)
print(f"Final Probability of Heads: {probs[-1]:.4f}")
```
