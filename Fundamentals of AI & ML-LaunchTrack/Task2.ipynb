```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/house_price_prediction.csv')
print("First 5 rows of the data:")
print(df.head())
print("\nData Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df['avg_income'] = df['avg_income'].fillna(df['avg_income'].mean())
df['avg_area_house_age'] = df['avg_area_house_age'].fillna(df['avg_area_house_age'].mean())
df['avg_area_num_rooms'] = df['avg_area_num_rooms'].fillna(df['avg_area_num_rooms'].mean())
df['avg_population'] = df['avg_population'].fillna(df['avg_population'].mean())

df.dropna(subset=['address'], inplace=True)
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
new_columns = {
    'avg_income': 'Income',
    'avg_area_house_age': 'House_Age',
    'avg_area_num_rooms': 'Num_Rooms',
    'avg_bedrooms': 'Bedrooms',
    'avg_population': 'Population',
    'price': 'Price',
    'address': 'Address'
}

df.rename(columns=new_columns, inplace=True)
print("New Column Names:")
print(df.columns)
```
